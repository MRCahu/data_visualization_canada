```python

df = pd.read_csv('../data/imigrantes_canada.csv')
df.info()
```

```python
df.set_index('Pa√≠s', inplace=True)
anos = list(map(str, range(1980, 2014)))
```

```python
brasil = df.loc['Brasil', anos]
dados_brasil = pd.DataFrame({
    'ano': list(map(int, anos)),
    'imigrantes': brasil.values
})
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'], marker='o')
plt.title('Imigra√ß√£o do Brasil para o Canad√° (1980‚Äì2013)')
plt.xlabel('Ano')
plt.ylabel('N√∫mero de Imigrantes')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

fig = px.line(dados_brasil, x='ano', y='imigrantes',
              title='Imigra√ß√£o do Brasil para o Canad√° (1980‚Äì2013)',
              markers=True)
fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

anos = list(map(int, range(1980, 2014)))

# Supondo que df tenha os dados com pa√≠ses no √≠ndice
brasil = df.loc['Brasil', list(map(str, anos))]
argentina = df.loc['Argentina', list(map(str, anos))]

# Criar figura com 1 linha e 2 colunas de gr√°ficos
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Gr√°fico Brasil
ax1.plot(anos, brasil, marker='o', color='blue')
ax1.set_title('Imigra√ß√£o do Brasil')
ax1.set_xlabel('Ano')
ax1.set_ylabel('N√∫mero de Imigrantes')
ax1.xaxis.set_major_locator(plt.MultipleLocator(5))
ax1.grid(True)

# Gr√°fico Argentina
ax2.plot(anos, argentina, marker='s', color='green')
ax2.set_title('Imigra√ß√£o da Argentina')
ax2.set_xlabel('Ano')
ax2.set_ylabel('N√∫mero de Imigrantes')
ax2.xaxis.set_major_locator(plt.MultipleLocator(5))
ax2.grid(True)

plt.suptitle('Imigra√ß√£o para o Canad√° (1980‚Äì2013): Brasil vs Argentina', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Exemplo de dados de imigra√ß√£o do Brasil
anos = list(range(1980, 2014))
imigrantes_brasil = df.loc['Brasil', list(map(str, anos))]

# Criar figura e eixo
fig, ax = plt.subplots(figsize=(10, 6))

# Plotar gr√°fico
ax.plot(anos, imigrantes_brasil, marker='o', color='blue')

# T√≠tulo com quebra de linha
ax.set_title("Imigra√ß√£o do Brasil para o Canad√°\n(1980 a 2013)", fontsize=14)

# R√≥tulos dos eixos
ax.set_xlabel("Ano")
ax.set_ylabel("N√∫mero de Imigrantes")

# Ticks personalizados
ax.xaxis.set_major_locator(plt.MultipleLocator(5))

# Grade
ax.grid(True)

# Exibir
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(anos, imigrantes_brasil, color='blue', label='Tend√™ncia')
ax.bar(anos, imigrantes_brasil, alpha=0.3, label='Valores por ano')
ax.set_title("Tend√™ncia e Total de Imigrantes\nBrasil para Canad√° (1980‚Äì2013)")
ax.set_xlabel("Ano")
ax.set_ylabel("Imigrantes")
ax.legend()
ax.grid(True)
plt.tight_layout()
plt.show()
```

```python
# prompt: SALVAR ESSE √öLTIMO GR√ÅFICO

# O √∫ltimo gr√°fico plotado foi o que mostra a tend√™ncia e os valores por ano.
# Para salvar o √∫ltimo gr√°fico, voc√™ pode adicionar plt.savefig()
# antes de plt.show() no √∫ltimo bloco de c√≥digo.

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(anos, imigrantes_brasil, color='blue', label='Tend√™ncia')
ax.bar(anos, imigrantes_brasil, alpha=0.3, label='Valores por ano')
ax.set_title("Tend√™ncia e Total de Imigrantes\nBrasil para Canad√° (1980‚Äì2013)")
ax.set_xlabel("Ano")
ax.set_ylabel("Imigrantes")
ax.legend()
ax.grid(True)
plt.tight_layout()

# Salva o √∫ltimo gr√°fico antes de exibi-lo
plt.savefig("tendencia_total_imigracao_brasil.png", dpi=300)

plt.show()
```
